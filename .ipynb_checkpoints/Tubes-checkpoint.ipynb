# Tugas Besar Pengolahan Citra Digial Kelompok 15
### Memanggil library numpy, opencv, dan matplotlib

```python
import numpy as np #import library numpy
import cv2 #import library opencv
import matplotlib.pyplot as plt #import library matplotlib
import math #import library math digunakan untuk memanggil method / fungsi sqrt()
```

### Membaca gambar menggunakan library opencv yang nantinya akan dilakukan pengolahan citra

```python
gbr1 = cv2.imread('1.jpg') #Untuk membaca Gambar
gbr2 = cv2.imread('2.jpg')
gbr3 = cv2.imread('3.jpg')
gbr4 = cv2.imread('4.jpg')
gbr5 = cv2.imread('5.jpg')
gbr6 = cv2.imread('1.jpg')

rgb1 = cv2.cvtColor(gbr1, cv2.COLOR_BGR2RGB) #cvtColor untuk mengkonversi gambar dari 1 color space ke color space lainnya
rgb2 = cv2.cvtColor(gbr2, cv2.COLOR_BGR2RGB) #Color_BGR2RGB untuk mengubah gambar dari yang BGR menjadi RGB, keduanya sama-sama rgb
rgb3 = cv2.cvtColor(gbr3, cv2.COLOR_BGR2RGB) #Dilakukan pengubahan ke RGB hanya untuk mempermudah, karena default dari OpenCV gambar berwarna dibaca dalam format BGR, walaupun sama-sama RGB
rgb4 = cv2.cvtColor(gbr4, cv2.COLOR_BGR2RGB) 
rgb5 = cv2.cvtColor(gbr5, cv2.COLOR_BGR2RGB) 
rgb6 = cv2.cvtColor(gbr6, cv2.COLOR_BGR2RGB) 

plt.title("Gambar Asli") #Memberikan judul untuk gambar
plt.imshow(rgb1) #Menampilkan data array sebagai gambar
plt.show() #Menampilkan figure/gambar yang ada pada imshow
plt.imshow(rgb2)
plt.show()
plt.imshow(rgb3)
plt.show()
plt.imshow(rgb4)
plt.show()
plt.imshow(rgb5)
plt.show()
plt.imshow(rgb6)
plt.show()
```

### Mengubah gambar menjadi gambar grayscale

```python
gray1 = cv2.cvtColor(rgb1, cv2.COLOR_RGB2GRAY) #RGB2GRAY mengubah gambar berwarna menjadi gambar grayscale
gray2 = cv2.cvtColor(rgb2, cv2.COLOR_RGB2GRAY)
gray3 = cv2.cvtColor(rgb3, cv2.COLOR_RGB2GRAY)
gray4 = cv2.cvtColor(rgb4, cv2.COLOR_RGB2GRAY)
gray5 = cv2.cvtColor(rgb5, cv2.COLOR_RGB2GRAY)
gray6 = cv2.cvtColor(rgb6, cv2.COLOR_RGB2GRAY)

plt.title("Gambar Grayscale") 
plt.imshow(gray1,cmap="gray")
plt.show()
plt.imshow(gray2,cmap="gray")
plt.show()
plt.imshow(gray3,cmap="gray")
plt.show()
plt.imshow(gray4,cmap="gray")
plt.show()
plt.imshow(gray5,cmap="gray")
plt.show()
plt.imshow(gray6,cmap="gray")
plt.show()
```

### Melakukan metode pelembutan citra (Image Smoothing)
Dilakukan metode pelembutan citra karena banyaknya bagian yang kasar pada gambar kanker kulit ganas (Skin Cancer Malignant)

```python
kernelq= np.array([[1/9, 1/9, 1/9], 
                   [1/9, 1/9, 1/9],
                   [1/9, 1/9, 1/9]])

#Mendeklarasikan array sebagai kernel untuk Pelembutan citra

hpf1 = cv2.filter2D(gray1, 0, kernelq) #Untuk melakukan konvolusi dengan kernel filtering
hpf2 = cv2.filter2D(gray2, 0, kernelq)
hpf3 = cv2.filter2D(gray3, 0, kernelq)
hpf4 = cv2.filter2D(gray4, 0, kernelq)
hpf5 = cv2.filter2D(gray5, 0, kernelq)
hpf6 = cv2.filter2D(gray6, 0, kernelq)

plt.title("Pelembutan Citra")
plt.imshow(hpf1,cmap="gray")
plt.show()
plt.imshow(hpf2,cmap="gray")
plt.show()
plt.imshow(hpf3,cmap="gray")
plt.show()
plt.imshow(hpf4,cmap="gray")
plt.show()
plt.imshow(hpf5,cmap="gray")
plt.show()
plt.imshow(hpf6,cmap="gray")
plt.show()
```

### Melakukan metode threshing
Dilakukan metode threshing yaitu untuk mengubah gambar grayscale menjadi gambar biner (Hitam dan putih).

```python
ret, thr1 = cv2.threshold(hpf1,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret, thr2 = cv2.threshold(hpf2,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret, thr3 = cv2.threshold(hpf3,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret, thr4 = cv2.threshold(hpf4,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret, thr5 = cv2.threshold(hpf5,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret, thr6 = cv2.threshold(hpf6,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

thr1=255-thr1
thr2=255-thr2
thr3=255-thr3
thr4=255-thr4
thr5=255-thr5
thr6=255-thr6

plt.title("Gambar Biner") 
plt.imshow(thr1,cmap="gray") 
plt.show()
plt.imshow(thr2,cmap="gray") 
plt.show()
plt.imshow(thr3,cmap="gray") 
plt.show()
plt.imshow(thr4,cmap="gray") 
plt.show()
plt.imshow(thr5,cmap="gray") 
plt.show()
plt.imshow(thr6,cmap="gray") 
plt.show()
```

### Melakukan metode operasi Erosi
Digunakan metode operasi erosi dikarenakan pada bagian kankernya digambar biner terlalu besar daripada gambar aslinya

```python
kernel_ero = np.ones((3,3),np.uint8)

ero1 = cv2.erode(thr1,kernel_ero,iterations=1)
ero2 = cv2.erode(thr2,kernel_ero,iterations=1)
ero3 = cv2.erode(thr3,kernel_ero,iterations=1)
ero4 = cv2.erode(thr4,kernel_ero,iterations=1)
ero5 = cv2.erode(thr5,kernel_ero,iterations=1)
ero6 = cv2.erode(thr6,kernel_ero,iterations=1)

plt.title("Gambar Hasil Operasi Erosi") 
plt.imshow(ero1,cmap="gray") 
plt.show()
plt.imshow(ero2,cmap="gray") 
plt.show()
plt.imshow(ero3,cmap="gray") 
plt.show()
plt.imshow(ero4,cmap="gray") 
plt.show()
plt.imshow(ero5,cmap="gray") 
plt.show()
plt.imshow(ero6,cmap="gray") 
plt.show()
```

### Melakukan metode operasi Opening
Digunakan metode operasi opening yaitu operasi erosi diikuti dengan operasi dilasi, operasi opening dilakukan untuk 
menghilangkan objek-objek kecil atau dapat dibilang memperhalus batas pada bagian kanker kulit ganas (Skin Cancer Malignant)

```python
kernel_dil = np.ones((3,3),np.uint8) #np.uint8 memberikan array tersebut data tipe berupa int 8 bit, dengan nilai maksimal 127 dan minimum -128

dil1 = cv2.erode(ero1,kernel_dil,iterations=1)
dil2 = cv2.erode(ero2,kernel_dil,iterations=1)
dil3 = cv2.erode(ero3,kernel_dil,iterations=1)
dil4 = cv2.erode(ero4,kernel_dil,iterations=1)
dil5 = cv2.erode(ero5,kernel_dil,iterations=1)
dil6 = cv2.erode(ero6,kernel_dil,iterations=1)

plt.title("Gambar Hasil Operasi Opening") 
plt.imshow(dil1,cmap="gray") 
plt.show()
plt.imshow(dil2,cmap="gray") 
plt.show()
plt.imshow(dil3,cmap="gray") 
plt.show()
plt.imshow(dil4,cmap="gray") 
plt.show()
plt.imshow(dil5,cmap="gray") 
plt.show()
plt.imshow(dil6,cmap="gray") 
plt.show()
```

### Melakukan metode Deteksi Tepi (Edge Detection)
Digunakan metode deteksi tepi untuk melihat pinggiran/tepi dari gambar kanker kulit ganas (Skin Cancer Malignant)

```python
#Membuat fungsi/method untuk deteksi tepi menggunakan operasi Prewitt
def prewitt(rgb):
    gray= rgb 
    height=np.size(rgb,0) 
    width=np.size(rgb,1) 
    
    sx  = np.array([[-1,0,1],  
                    [-1,0,1],
                    [-1,0,1]])
    
    sy  = np.array([[1,1,1], 
                    [0,0,0],
                    [-1,-1,-1]])
    
    zxc=3 
    pad=zxc-1 
    jarak=int(pad/2) 

    tampung=np.zeros(gray.shape) 
    low = np.zeros((height+pad,width+pad))

    for i in range(height):
        for y in range(width):
            low[i+jarak,y+jarak]=gray[i,y] 

    for i in range (height):
        for y in range(width):
            nSx=0 
            nSy=0 
            
            for q in range(zxc):
                for w in range(zxc):
                    nSx+=low[i+q,y+w]*sx[q,w] 
                    nSx+=low[i+q,y+w]*sy[q,w]
                    
            hasil=round(math.sqrt((nSx**2)+(nSy**2))) 
            tampung[i,y]=hasil
    
    plt.imshow(tampung,cmap="gray") 
    plt.show()
```

```python
plt.title("Deteksi Tepi")
prewitt(dil1)
prewitt(dil2)
prewitt(dil3)
prewitt(dil4)
prewitt(dil5)
prewitt(dil6)
```
